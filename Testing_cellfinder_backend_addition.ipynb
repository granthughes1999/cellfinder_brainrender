## Create randomized values for all the columns of the tdTomato_df, for testing the statsitcal analysis

```python
# make each column value equal to zero
tdTomato_df = tdTomato_df.assign(total_cells=0)
tdTomato_df = tdTomato_df.assign(left_cell_count=0)
tdTomato_df = tdTomato_df.assign(right_cell_count=0)
tdTomato_df = tdTomato_df.assign(left_volume_mm3=0)
tdTomato_df = tdTomato_df.assign(right_volume_mm3=0)
tdTomato_df = tdTomato_df.assign(left_cells_per_mm3	=0)
tdTomato_df = tdTomato_df.assign(right_cells_per_mm3=0)

# Generate an array of random integers that are between relavent int for that column
random_values_totalcells = np.random.randint(9000, 15000, size=len(tdTomato_df))
random_values_leftcells = np.random.randint(9000, 15000, size=len(tdTomato_df))
random_values_rightcells = np.random.randint(9000, 15000, size=len(tdTomato_df))
random_values_leftvolume = np.random.randint(1, 10, size=len(tdTomato_df))
random_values_rightvolume = np.random.randint(1, 10, size=len(tdTomato_df))
random_values_leftpermm3 = np.random.randint(750, 7000, size=len(tdTomato_df))
random_values_rightpermm3 = np.random.randint(750, 7000, size=len(tdTomato_df))

# Add the random values to the values to the column
tdTomato_df['total_cells'] = tdTomato_df['total_cells'] + random_values_totalcells
tdTomato_df['left_cell_count'] = tdTomato_df['left_cell_count'] + random_values_leftcells
tdTomato_df['right_cell_count'] = tdTomato_df['right_cell_count'] + random_values_rightcells
tdTomato_df['left_volume_mm3'] = tdTomato_df['left_volume_mm3'] + random_values_leftvolume
tdTomato_df['right_volume_mm3'] = tdTomato_df['right_volume_mm3'] + random_values_rightvolume
tdTomato_df['left_cells_per_mm3'] = tdTomato_df['left_cells_per_mm3'] + random_values_leftpermm3
tdTomato_df['right_cells_per_mm3'] = tdTomato_df['right_cells_per_mm3'] + random_values_rightpermm3

randomized_tdTomato_df = tdTomato_df
randomized_tdTomato_df.to_csv(cellfinder_output_path + 'analysis/' + 'randomized_tdTomato_summary.csv')
randomized_tdTomato_df

```

### divide one column value against another, add that value to a new column. thanks chatGPT...

```python
# Use the assign method to create a new column 'C' and perform the division operation
labled_cells_df = labled_cells_df.assign(percent_cell_count_left=gfp_df['left_cell_count']/tdTomato_df['left_cell_count']*100)
labled_cells_df = labled_cells_df.assign(percent_cell_count_right=gfp_df['right_cell_count']/tdTomato_df['right_cell_count']*100)

labled_cells_df = labled_cells_df.assign(percent_volume_per_mm3_left=gfp_df['right_volume_mm3']/tdTomato_df['right_volume_mm3']*100)
labled_cells_df = labled_cells_df.assign(percent_percent_volume_per_mm3_right=gfp_df['left_volume_mm3']/tdTomato_df['left_volume_mm3']*100)

labled_cells_df = labled_cells_df.assign(percent_cells_per_mm3_right=gfp_df['right_cells_per_mm3']/tdTomato_df['right_cells_per_mm3']*100)
labled_cells_df = labled_cells_df.assign(percent_cells_per_mm3_left=gfp_df['left_cells_per_mm3']/tdTomato_df['left_cells_per_mm3']*100)

labled_cells_df
```

### These next cells add 2 columns to the labled_cells df, that contain the percent of gfp labled cells in the left and right hemispsheres 

```python
# make list of cell count for each brainregin in left hemi for tdTomato
tdTomato_left_cells = tdTomato_df['left_cell_count'].to_list()

# make list of cell count for each brainregin in left hemi for GFP
gfp_left_cells = gfp_df['left_cell_count'].to_list()

# make list of cell count for each brainregin in right hemi for tdTomato
tdTomato_right_cells = tdTomato_df['right_cell_count'].to_list()

# make list of cell count for each brainregin in right hemi for GFP
gfp_right_cells = gfp_df ['right_cell_count'].to_list()
```

```python
# making a df with percent of gfp cells labled compared to tdTomato
percent_labled_left = []
i = 0
for count in tdTomato_left_cells:
    if gfp_left_cells[i] >=1:
        percent_labled_left.append(gfp_left_cells[i] / count * 100)
        i += 1
    else:
        percent_labled_left.append(0)
        i += 1

print(len(percent_labled_left))


# making a df with percent of gfp cells labled compared to tdTomato
percent_labled_right = []
i = 0
for count in tdTomato_right_cells:
    if gfp_right_cells[i] >=1:
        percent_labled_right.append(gfp_right_cells[i] / count * 100)
        i += 1
    else:
        percent_labled_right.append(0)
        i += 1
```

```python
percent_labled_left_matched_len = []
percent_labled_right_matched_len = []
for i in range(len(labled_cells_df)):
    percent_labled_left_matched_len.append(percent_labled_left[i])
    percent_labled_right_matched_len.append(percent_labled_right[i])

# create df with gfp and tdTomato cell counts, and percent labled
labled_cells_df['percent of cells in left hemi gfp/tdTomato'] = percent_labled_left_matched_len
labled_cells_df['percent of cells in right hemi gfp/tdTomato'] = percent_labled_right_matched_len

# labled_cells_df.to_csv(output_path_csv)
labled_cells_df
```
